```python
import shapefile
```

```python
with shapefile.Writer('out/rm_bbox') as shp:
    shp.field('name', 'C')
    for i in range(len(rm_bbox)):
        x0 = rm_bbox.x0_geo.iloc[i]
        y0 = rm_bbox.y0_geo.iloc[i]
        x1 = rm_bbox.x1_geo.iloc[i]
        y1 = rm_bbox.y1_geo.iloc[i]
        
        coord = [[x0, y0],[x1, y0],[x1, y1],[x0, y1],[x0, y0]]
        shp.poly([coord])
        shp.record(str(i))
```

```python
def save_shp(bbox, shp_path):
    with shapefile.Writer(shp_path) as shp:
        shp.field('name', 'C')
        for i in range(len(bbox)):
            x0 = bbox.x0_geo.iloc[i]
            y0 = bbox.y0_geo.iloc[i]
            x1 = bbox.x1_geo.iloc[i]
            y1 = bbox.y1_geo.iloc[i]

            coord = [[x0, y0],[x1, y0],[x1, y1],[x0, y1],[x0, y0]]
            shp.poly([coord])
            shp.record(str(i))
```

```python
save_shp(keep_bbox, 'out/keep_bbox')
```
