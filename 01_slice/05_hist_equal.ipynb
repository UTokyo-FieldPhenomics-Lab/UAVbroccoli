```python
import os
import skimage.io
import numpy as np
import matplotlib.pyplot as plt
```

```python
img_folder = f"Z:/共享云端硬盘/broccoliProject/10_anotation_use/png/broccoli_tanashi_5_20200417_P4RGB_15m_M"
out_folder = r"Z:\共享云端硬盘\broccoliProject\10_locate_by_cv\hist_equal_0417"
```

```python
for img_name in os.listdir(img_folder):
    img_path = os.path.join(img_folder, img_name)
    break
```

```python
test_img = skimage.io.imread(img_path)

plt.imshow(test_img)
```

```python
test_img.shape
```

```python
img_eq = skimage.exposure.equalize_hist(test_img)
plt.imshow(img_eq)
```

```python
img_eq = skimage.exposure.equalize_hist(test_img, mask=test_img[:,:,3] == 255)

plt.imshow(img_eq)
```

```python
for img_name in os.listdir(img_folder):
    img_path = os.path.join(img_folder, img_name)
    test_img = skimage.io.imread(img_path)
    
    img_eq = skimage.exposure.equalize_hist(test_img, mask=test_img[:,:,3] == 255)
    
    out_img_path = os.path.join(out_folder, img_name)
    skimage.io.imsave(out_img_path, (img_eq*255).astype(np.uint8))
```
