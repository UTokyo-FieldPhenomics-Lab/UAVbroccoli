```python
root20 = r"Y:\hwang_Pro\data\2020_tanashi_broccoli\13_roi_on_raw"
root21 = r"Y:\hwang_Pro\data\2021_tanashi_broccoli\13_roi_on_raw"
```

```python
date20 = ["0518_p", "0520_p", "0522_p", "0525_p", "0526_p", "0528_p"]
date21 = ["210512", "210514", "210515", "210519", "210520", "210526"]
```

```python
verify_list = os.listdir(f"{root20}/verify")
verify_list
```

```python
csv_file = pd.DataFrame(columns=["date", "file_name", "relative_path", "status"])
for d in date20:
    p2 = Paths(d)
    json_name = f"{root20}/{p2.project_name}.json"
    date = p2.project_name.split("_")[3]
    with open(json_name, 'r') as f:
        offset_dict = json.load(f)
    for img_name in offset_dict.keys():
        if f"labeled_{date}_{img_name[:-4]}.json" in verify_list:
            status = "verify"
        else:
            status = "blank"
        csv_file.loc[len(csv_file)] = [date, img_name, offset_dict[img_name]["imagePath"], status]
```

```python
csv_file.to_csv(f"{root20}\json_index.csv", index=False)
```
